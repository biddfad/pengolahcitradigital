```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

# --- 1. ALGORITMA RLE (CORE) ---
def rle_compression(data):
    """Mengubah list data menjadi list tuple (nilai, jumlah)"""
    if not data: return []
    
    compressed = []
    count = 1
    for i in range(1, len(data)):
        if data[i] == data[i-1]:
            count += 1
        else:
            compressed.append((data[i-1], count))
            count = 1
    # Jangan lupa elemen terakhir
    compressed.append((data[-1], count)) 
    return compressed

def rle_decompression(compressed):
    """Mengembalikan list tuple (nilai, jumlah) menjadi list data asli"""
    decompressed = []
    for value, count in compressed:
        decompressed.extend([value] * count)
    return decompressed

# --- 2. FUNGSI PENGOLAHAN GAMBAR ---
def proses_gambar_rle(nama_file):
    print(f"--- Memproses: {nama_file} ---")
    
    # A. Buka Gambar
    try:
        img_asli = Image.open(nama_file)
    except FileNotFoundError:
        print(f"Error: File '{nama_file}' tidak ditemukan. Pastikan sudah di-upload.")
        return

    # B. Pre-processing (Resize & Quantization)
    # Kita ubah jadi 100x100 dan batasi 16 warna agar RLE efektif
    img = img_asli.resize((100, 100)) 
    img = img.convert('P', palette=Image.ADAPTIVE, colors=16) 
    
    width, height = img.size
    
    # C. Proses Kompresi
    pixels = list(img.getdata()) # Ratakan jadi 1 dimensi
    compressed_data = rle_compression(pixels)
    
    # Hitung Efisiensi
    ukuran_asli = len(pixels)
    ukuran_kompresi = len(compressed_data)
    ratio = (1 - (ukuran_kompresi / ukuran_asli)) * 100
    print(f"Jumlah Data Piksel Asli : {ukuran_asli}")
    print(f"Jumlah Data Terkompresi : {ukuran_kompresi}")
    print(f"Efisiensi Kompresi      : {ratio:.2f}%")

    # D. Proses Dekompresi
    decompressed_pixels = rle_decompression(compressed_data)
    
    # E. Rekonstruksi Gambar (MENYUSUN ULANG)
    img_hasil = Image.new(img.mode, (width, height))
    img_hasil.putdata(decompressed_pixels)
    
    # === PERBAIKAN PENTING (THE FIX) ===
    # Menyalin palet warna dari gambar asli ke gambar hasil dekompresi
    # Tanpa baris ini, gambar akan terlihat hitam karena komputer tidak tahu arti angka warnanya.
    img_hasil.putpalette(img.getpalette()) 
    # ===================================

    # F. Visualisasi Hasil (Matplotlib)
    plt.figure(figsize=(10, 5))

    # Plot 1: Gambar Asli (yang sudah diproses ukurannya)
    plt.subplot(1, 2, 1)
    plt.title(f"Original (16 Warna)\nSize: {width}x{height}")
    plt.imshow(img)
    plt.axis('off')

    # Plot 2: Hasil Dekompresi
    plt.subplot(1, 2, 2)
    plt.title("Hasil Dekompresi RLE\n(Warna Sudah Kembali)")
    plt.imshow(img_hasil)
    plt.axis('off')

    plt.tight_layout()
    plt.show()

# --- 3. EKSEKUSI ---
# Pastikan nama file sesuai dengan yang kamu upload
proses_gambar_rle('kiwi.jpg')
```

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

# --- 1. ALGORITMA RLE ---
def rle_compression(data):
    if not data: return []
    compressed = []
    count = 1
    for i in range(1, len(data)):
        if data[i] == data[i-1]:
            count += 1
        else:
            compressed.append((data[i-1], count))
            count = 1
    compressed.append((data[-1], count)) 
    return compressed

def rle_decompression(compressed):
    decompressed = []
    for value, count in compressed:
        decompressed.extend([value] * count)
    return decompressed

# --- 2. PROSES UTAMA ---
def proses_gambar_biner(nama_file):
    print(f"--- Memproses: {nama_file} ---")
    
    # A. Buka Gambar Asli
    try:
        img_asli = Image.open(nama_file)
    except FileNotFoundError:
        print("File tidak ditemukan.")
        return

    # B. Ubah ke Hitam Putih (Thresholding)
    # Langkah 1: Ubah ke Grayscale (Abu-abu)
    img_gray = img_asli.convert('L')
    
    # Langkah 2: Tentukan batas (Threshold)
    # Angka < 128 jadi 0 (Hitam), Angka > 128 jadi 255 (Putih)
    # Ini membuat gambar jadi siluet tegas.
    threshold_value = 128
    img_biner = img_gray.point(lambda x: 255 if x > threshold_value else 0, mode='1')
    
    width, height = img_biner.size
    
    # C. Kompresi RLE
    # Kita ambil datanya. Karena mode '1', isinya cuma 0 atau 1 (atau 255)
    pixels = list(img_biner.getdata())
    
    compressed_data = rle_compression(pixels)
    
    # Hitung Efisiensi
    ukuran_asli = len(pixels)
    ukuran_kompresi = len(compressed_data)
    # Rumus efisiensi: seberapa banyak data yang "hilang/dihemat"
    ratio = (1 - (ukuran_kompresi / ukuran_asli)) * 100
    
    print(f"Piksel Asli      : {ukuran_asli}")
    print(f"Jumlah RLE Tuple : {ukuran_kompresi}")
    print(f"Efisiensi        : {ratio:.2f}% (Sangat Tinggi!)")

    # D. Dekompresi
    decompressed_pixels = rle_decompression(compressed_data)
    
    # E. Susun Kembali
    img_hasil = Image.new(img_biner.mode, (width, height))
    img_hasil.putdata(decompressed_pixels)
    
    # F. Tampilkan
    plt.figure(figsize=(10, 5))

    # Gambar 1: Asli (Warna)
    plt.subplot(1, 3, 1)
    plt.title("Gambar Asli")
    plt.imshow(img_asli)
    plt.axis('off')

    # Gambar 2: Input RLE (Biner)
    plt.subplot(1, 3, 2)
    plt.title("Input Biner\n(Thresholding)")
    plt.imshow(img_biner) # Matplotlib otomatis tahu ini hitam putih
    plt.axis('off')

    # Gambar 3: Output RLE
    plt.subplot(1, 3, 3)
    plt.title(f"Hasil Dekompresi\n(Sama Persis Input Biner)")
    plt.imshow(img_hasil)
    plt.axis('off')

    plt.tight_layout()
    plt.show()

# Jalankan dengan file kiwi kamu
proses_gambar_biner('kiwi.jpg')
```

```python
# --- STEP 1: SEGMENTASI CITRA (Memisahkan Buah dari Background) ---

nama_file = 'kiwi.jpg'  # Pastikan file sudah di-upload
img = cv2.imread(nama_file)

if img is None:
    print("Error: Gambar tidak ditemukan!")
else:
    # 1. Resize agar ringan (Opsional, 100x100 cukup buat demo)
    img = cv2.resize(img, (100, 100))

    # 2. Konversi ke HSV (Hue Saturation Value)
    # HSV lebih jago membedakan warna daripada RGB
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    # 3. Tentukan Rentang Warna HIJAU (Segmentasi Warna)
    # Rentang ini khusus untuk buah Kiwi hijau
    lower_green = np.array([25, 40, 40])
    upper_green = np.array([90, 255, 255])

    # 4. Buat Masking
    # Ini akan membuat area hijau jadi PUTIH, sisanya HITAM
    mask = cv2.inRange(hsv, lower_green, upper_green)

    # 5. Terapkan Mask ke Gambar Asli (Ekstraksi)
    # Latar belakang dibuang jadi hitam pekat (Nilai 0)
    # Variabel ini yang nanti akan dikirim ke RLE
    img_segmentasi = cv2.bitwise_and(img, img, mask=mask)

    # --- Tampilkan Hasil Segmentasi ---
    plt.figure(figsize=(10, 4))

    # Gambar Asli
    plt.subplot(1, 2, 1)
    plt.title("Gambar Asli")
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')

    # Hasil Segmentasi
    plt.subplot(1, 2, 2)
    plt.title("Hasil Segmentasi\n(Siap untuk RLE)")
    plt.imshow(cv2.cvtColor(img_segmentasi, cv2.COLOR_BGR2RGB))
    plt.axis('off')

    plt.show()

    print("Cell 2 Berhasil: Gambar sudah disegmentasi. Background sudah bersih!")
```

```python
%%writefile app.py
import streamlit as st
import cv2
import numpy as np
from PIL import Image

# --- 1. FUNGSI LOGIKA (BACKEND) ---

def rle_compression(data):
    if len(data) == 0: return []
    compressed = []
    count = 1
    for i in range(1, len(data)):
        if data[i] == data[i-1]:
            count += 1
        else:
            compressed.append((data[i-1], count))
            count = 1
    compressed.append((data[-1], count))
    return compressed

def rle_decompression(compressed):
    decompressed = []
    for value, count in compressed:
        decompressed.extend([value] * count)
    return decompressed

def proses_segmentasi(image_array):
    # Resize agar proses cepat
    img = cv2.resize(image_array, (200, 200)) # Kita set 200x200 untuk demo
    
    # Convert ke HSV untuk deteksi warna
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    
    # Range Warna Hijau (Sesuaikan jika buahnya bukan Kiwi/Apel Hijau)
    lower_green = np.array([25, 40, 40])
    upper_green = np.array([90, 255, 255])
    
    # Masking
    mask = cv2.inRange(hsv, lower_green, upper_green)
    
    # Ekstraksi (Background jadi Hitam)
    result = cv2.bitwise_and(img, img, mask=mask)
    return img, result

# --- 2. TAMPILAN APLIKASI (FRONTEND) ---

st.set_page_config(page_title="RLE Image Compressor", layout="wide")

st.title(" Mini App: Kompresi Citra RLE")
st.write("Aplikasi sederhana untuk segmentasi objek dan kompresi Run-Length Encoding.")

# -- Sidebar untuk Upload --
st.sidebar.header("1. Upload Gambar")
uploaded_file = st.sidebar.file_uploader("Pilih gambar buah (Format JPG/PNG)", type=["jpg", "png", "jpeg"])

if uploaded_file is not None:
    # Baca file gambar dari upload
    file_bytes = np.asarray(bytearray(uploaded_file.read()), dtype=np.uint8)
    img_cv = cv2.imdecode(file_bytes, 1)
    
    # --- PROSES 1: SEGMENTASI ---
    st.header("Step 1: Segmentasi & Preprocessing")
    
    col1, col2 = st.columns(2)
    
    # Proses Segmentasi
    img_asli, img_segmentasi = proses_segmentasi(img_cv)
    
    # Konversi BGR (OpenCV) ke RGB (Streamlit/Layar)
    img_asli_show = cv2.cvtColor(img_asli, cv2.COLOR_BGR2RGB)
    img_seg_show = cv2.cvtColor(img_segmentasi, cv2.COLOR_BGR2RGB)
    
    with col1:
        st.image(img_asli_show, caption="Gambar Asli (Resized)", use_container_width=True)
    with col2:
        st.image(img_seg_show, caption="Hasil Segmentasi (Background Hitam)", use_container_width=True)
        st.info("Background dihapus agar kompresi RLE efisien.")

    # --- PROSES 2: KOMPRESI RLE ---
    st.header("Step 2: Kompresi RLE")
    
    if st.button("Lakukan Kompresi RLE"):
        with st.spinner('Sedang memproses...'):
            # Persiapan Data
            pixels = img_seg_show.flatten().tolist()
            
            # Eksekusi RLE
            compressed_data = rle_compression(pixels)
            
            # Hitung Statistik
            ukuran_asli = len(pixels)
            ukuran_kompresi = len(compressed_data)
            rasio = (1 - (ukuran_kompresi / ukuran_asli)) * 100
            
            # Tampilkan Statistik dengan Metrik Keren
            col_stat1, col_stat2, col_stat3 = st.columns(3)
            col_stat1.metric("Total Piksel Asli", f"{ukuran_asli:,}")
            col_stat2.metric("Ukuran RLE (Tuple)", f"{ukuran_kompresi:,}")
            col_stat3.metric("Efisiensi Kompresi", f"{rasio:.2f}%")
            
            st.success("Kompresi Selesai!")
            
            # Intip Data
            st.text("Sampel Data Terkompresi (20 data pertama):")
            st.code(str(compressed_data[:20]), language='python')

            # --- PROSES 3: DEKOMPRESI ---
            st.header("Step 3: Dekompresi (Pembuktian)")
            
            decompressed_pixels = rle_decompression(compressed_data)
            
            # Reshape kembali ke bentuk gambar
            h, w, c = img_segmentasi.shape
            img_hasil = np.array(decompressed_pixels, dtype=np.uint8).reshape((h, w, c))
            
            st.image(img_hasil, caption="Hasil Dekompresi (Sama persis dengan input)", width=300)

else:
    st.info("Silakan upload gambar di menu sebelah kiri untuk memulai.")
```
